```python
import chess.pgn
pgn = open('GarryKasparov.pgn')
# https://chess-db.com/public/pinfo.jsp?id=4100018
```

```python
import numpy as np
def makeRows(num):
    pieceposition = []
    for i in range(num):
        pieceposition.append(np.zeros((8,8), dtype=int))
    return pieceposition
db = dict(
    p=makeRows(50),
    r=makeRows(50),
    n=makeRows(50),
    b=makeRows(50),
    q=makeRows(50),
    k=makeRows(50)
    )
```

```python
def getGame():
    game = chess.pgn.read_game(pgn)
    if game == None:
        return
    board = game.board()
    boardlist = []
    for move in game.main_line():
        board.push(move)
        bstring = board.__str__().lower().replace('\n','').replace(' ', '')
        boardlist.append(bstring)
    return boardlist
```

```python
def appendData(pieceType):
    move = 0
    for position in positions:
        if move >= 50:
            break
        instance = []
        for i in range(64):
            s = position
            if s[i:i+1] == pieceType:
                instance.append(i)
        for value in instance:
            file = 7 - (value // 8)
            rank = value % 8
            db[pieceType][move][file][rank] += 1
        move += 1
    return db
```

```python
while(True):
    try:
        positions = getGame()
        appendData('p')
        appendData('r')
        appendData('n')
        appendData('b')
        appendData('k')
        appendData('q')
        chess.pgn.read_game(pgn)
    except:
        break
```
